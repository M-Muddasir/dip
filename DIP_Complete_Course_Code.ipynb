# PART 1: Matrix & Array Operations

## 00_matrix.py - Matrix Operations (Deprecated Method)

Using `numpy.matrix` (not recommended anymore, shown for reference)

````python
# 00_matrix.py - Matrix operations using numpy.matrix (deprecated)
# This is just for reference to show that matrix exists in Python like MATLAB

from numpy import matrix, linalg as la

print("Two matrices as matrix objects")
a = matrix([[3, 1], [2, 5]])
b = matrix([[7, 9], [8, 5]])
print("a =")
print(a)
print("b =")
print(b)

print("\nBasic arithmetic operations on matrices (matrix type)")
m1 = a + b
m2 = a - b
m3 = a * b  # * is matrix multiplication for matrix type
m4 = a / b  # / is element-wise division

print("Matrix addition:\n", m1)
print("Matrix difference:\n", m2)
print("Matrix product (a * b):\n", m3)
print("Element-wise division:\n", m4)

print("\nTranspose and inverse of a matrix")
at = a.T
ai = la.inv(a)

print("Original matrix:\n", a)
print("Transpose:\n", at)
print("Inverse:\n", ai)

print("\nProduct of a matrix and its inverse")
m1 = a * ai
m2 = ai * a

print("A * A⁻¹:\n", m1)
print("A⁻¹ * A:\n", m2)
````

## 01_array.py - Array Operations (Recommended Method)

Using `numpy.array` - the modern and recommended approach

````python
# 01_array.py - Array operations using numpy.array (RECOMMENDED)

import numpy as np
from numpy import linalg as la

print("Two matrices as NumPy arrays")
a = np.array([[3, 1], [2, 5]])
b = np.array([[7, 9], [8, 5]])
print("a =")
print(a)
print("b =")
print(b)

print("\nBasic arithmetic operations on arrays")

m1 = a + b
m2 = a - b
m3 = a * b          # * is ELEMENT-WISE multiplication for arrays
m4 = a @ b          # @ is MATRIX multiplication for arrays
m5 = a / b          # / is element-wise division

print("Matrix addition:\n", m1)
print("Matrix difference:\n", m2)
print("Element-wise product (a * b):\n", m3)
print("Matrix product (a @ b):\n", m4)
print("Element-wise division:\n", m5)

print("\nTranspose and inverse")
at = a.T
ai = la.inv(a)

print("Original matrix:\n", a)
print("Transpose:\n", at)
print("Inverse:\n", ai)

print("\nProduct of a matrix and its inverse")
m1 = a @ ai
m2 = ai @ a

print("A @ A⁻¹:\n", m1)
print("A⁻¹ @ A:\n", m2)
````

### Key Difference: matrix vs array

| Operation | numpy.matrix | numpy.array |
|-----------|-------------|-------------|
| `*` | Matrix multiplication | Element-wise multiplication |
| `@` | Matrix multiplication | Matrix multiplication |
| Recommended | NO (deprecated) | YES |

# PART 5: Homogeneous Coordinates & Linear Algebra

## homogeneous.py - Solve Linear Equations & Null Space

````python
# homogeneous.py - Solving linear equations and null space

import numpy as np

# --- Part 1: Solve Ax = b ---
print("=== Part 1: Solve Ax = b ===")

A = np.array([[2, -3],
              [3,  5]], dtype=float)

b = np.array([[6],
              [10]], dtype=float)

print("A =")
print(A)
print("b =")
print(b)

# Equivalent to inv(A) * b
x = np.linalg.inv(A) @ b

print("\nSolution x = inv(A) @ b:")
print(x)
````

````python
# --- Part 2: Null space of B = [A | -b] ---
print("=== Part 2: Null Space (Homogeneous Coordinates) ===")

B = np.array([[2, -3, -6],
              [3,  5, -10]], dtype=float)

print("B = [A | -b] =")
print(B)

# Compute null space using SVD
U, S, Vt = np.linalg.svd(B)

# Null space vector corresponds to smallest singular value
o = Vt.T[:, -1]

print("\nNull space vector of B:")
print(o.reshape(-1, 1))

# Normalize by the last element (homogeneous to Cartesian)
o_normalized = o / o[-1]

print("\nNull space vector (normalized, homogeneous → Cartesian):")
print(o_normalized.reshape(-1, 1))
````

# PART 6: Interpolation

## li.py - Linear Interpolation (1D)

````python
# li.py - Linear Interpolation

def linear_interpolation(x, points):
    """
    Linear interpolation using two points.
    
    Formula: f(x) = f1 + (x - x1) * (f2 - f1) / (x2 - x1)
    
    points: [(x1, f1), (x2, f2)]
    x: location where interpolation is required
    """
    (x1, f1), (x2, f2) = points
    
    if x1 == x2:
        raise ValueError("x1 and x2 cannot be equal.")
    
    return f1 + (x - x1) * (f2 - f1) / (x2 - x1)


# Example usage
points = [
    (0, 10),   # f(0) = 10
    (1, 30)    # f(1) = 30
]

x = 0.5
value = linear_interpolation(x, points)

print(f"Given: f(0)=10, f(1)=30")
print(f"Interpolated value at x={x}: {value}")
````

## bli1.py - Bilinear Interpolation (Matrix Method)

````python
# bli1.py - Bilinear Interpolation using matrix solve

import numpy as np

def bilinear_interpolation_matrix(x, y, points):
    """
    Bilinear interpolation using solution of 4 linear equations.
    
    f(x,y) = a0 + a1*x + a2*y + a3*x*y
    
    points: [(x1,y1,f11), (x2,y1,f21), (x1,y2,f12), (x2,y2,f22)]
    """
    (x1, y1, f11), (x2, y1b, f21), (x1b, y2, f12), (x2b, y2b, f22) = points
    
    # Construct system matrix
    A = np.array([
        [1, x1, y1, x1*y1],
        [1, x2, y1, x2*y1],
        [1, x1, y2, x1*y2],
        [1, x2, y2, x2*y2]
    ])
    
    # Right-hand side vector
    b = np.array([f11, f21, f12, f22])
    
    # Solve for coefficients [a0, a1, a2, a3]
    coeff = np.linalg.solve(A, b)
    a0, a1, a2, a3 = coeff
    
    # Compute interpolated value
    return a0 + a1*x + a2*y + a3*x*y


# Example usage
points = [
    (0, 0, 10),   # f(0,0) = 10
    (1, 0, 20),   # f(1,0) = 20
    (0, 1, 30),   # f(0,1) = 30
    (1, 1, 40)    # f(1,1) = 40
]

x, y = 0.5, 0.5
value = bilinear_interpolation_matrix(x, y, points)

print(f"Given: f(0,0)=10, f(1,0)=20, f(0,1)=30, f(1,1)=40")
print(f"Interpolated value at ({x},{y}) = {value}")
````

## bli2.py - Bilinear Interpolation (Direct Formula)

````python
# bli2.py - Bilinear Interpolation using direct formula

def bilinear_interpolation_formula(x, y, points):
    """
    Bilinear interpolation using four corner points.
    
    Direct formula (weighted average):
    f(x,y) = [f11*(x2-x)*(y2-y) + f21*(x-x1)*(y2-y) +
              f12*(x2-x)*(y-y1) + f22*(x-x1)*(y-y1)] / [(x2-x1)*(y2-y1)]
    
    points: [(x1,y1,f11), (x2,y1,f21), (x1,y2,f12), (x2,y2,f22)]
    """
    (x1, y1, f11), (x2, y1b, f21), (x1b, y2, f12), (x2b, y2b, f22) = points
    
    # Bilinear interpolation formula
    return (
        f11 * (x2 - x) * (y2 - y) +
        f21 * (x - x1) * (y2 - y) +
        f12 * (x2 - x) * (y - y1) +
        f22 * (x - x1) * (y - y1)
    ) / ((x2 - x1) * (y2 - y1))


# Example usage
points = [
    (0, 0, 10),   # f(0,0) = 10
    (1, 0, 20),   # f(1,0) = 20
    (0, 1, 30),   # f(0,1) = 30
    (1, 1, 40)    # f(1,1) = 40
]

x, y = 0.5, 0.5
value = bilinear_interpolation_formula(x, y, points)

print(f"Given: f(0,0)=10, f(1,0)=20, f(0,1)=30, f(1,1)=40")
print(f"Interpolated value at ({x},{y}) = {value}")
````

### Interpolation Summary

| Type | Formula | Use Case |
|------|---------|----------|
| Linear | `f1 + (x-x1)*(f2-f1)/(x2-x1)` | 1D interpolation |
| Bilinear | Weighted average of 4 corners | 2D interpolation (images) |

# PART 7: Geometric Transformations

## Load Image for Transformations

````python
from PIL import Image
from numpy import array, zeros
from matplotlib.pyplot import imshow, show, subplot, figure, title

# Load image
img1 = Image.open('parrot.jpg').convert('L')
im1 = array(img1)
rows = im1.shape[0]
cols = im1.shape[1]

print(f"Image size: {rows} x {cols}")
imshow(im1, cmap='gray')
title('Original Image')
show()
````

## t1-scale.py - Scaling Transformation

````python
# t1-scale.py - Scale transformation
# Scale matrix (2x zoom):
# [sx  0 ]
# [0   sy]

sm = array([[2, 0],
            [0, 2]])  # Scale by 2x in both directions

print("Scale Matrix (2x):")
print(sm)

# Calculate output size
tp = sm @ array([[rows], [cols]])
rowsd = int(tp[0, 0]) + 1
colsd = int(tp[1, 0]) + 1

print(f"Output size: {rowsd} x {colsd}")

# Create output image
im2_scale = zeros([rowsd, colsd])

# Forward mapping
r = 0
while r < rows:
    c = 0
    while c < cols:
        tp = sm @ array([[r], [c]])
        rd = int(tp[0, 0])
        cd = int(tp[1, 0])
        im2_scale[rd, cd] = im1[r, c]
        c = c + 1
    r = r + 1

print("Scaling complete!")
````

````python
# Display scaling result
figure(figsize=(12, 5))
subplot(1, 2, 1)
imshow(im1, cmap='gray')
title('Original')
subplot(1, 2, 2)
imshow(im2_scale, cmap='gray')
title('Scaled 2x')
show()
````

## t2-rotation.py - Rotation Transformation (30°)

````python
# t2-rotation.py - Rotation transformation
# Rotation matrix (30 degrees):
# [cos θ   -sin θ   0]
# [sin θ    cos θ   0]
# [0        0       1]
#
# cos(30°) = 0.866, sin(30°) = 0.5

sm = [[.866, -.5, 0],
      [.5, .866, 0],
      [0, 0, 1]]  # Homogeneous rotation matrix

print("Rotation Matrix (30°):")
print(array(sm))

# Calculate output size
tp = sm @ array([[rows], [cols], [1]])
rowsd = int(tp[0, 0]) + 200
colsd = int(tp[1, 0]) + 200

# Create output image
im2_rotation = zeros([rowsd, colsd])

# Forward mapping
r = 0
while r < rows:
    c = 0
    while c < cols:
        tp = sm @ array([[r], [c], [1]])
        rd = int(tp[0, 0])
        cd = int(tp[1, 0])
        td = int(tp[2, 0])  # td is always 1 for affine
        if rd >= 0 and cd >= 0:
            im2_rotation[rd, cd] = im1[r, c]
        c = c + 1
    r = r + 1

print("Rotation complete!")
````

````python
# Display rotation result
figure(figsize=(12, 5))
subplot(1, 2, 1)
imshow(im1, cmap='gray')
title('Original')
subplot(1, 2, 2)
imshow(im2_rotation, cmap='gray')
title('Rotated 30°')
show()
````

## t3-projection.py - Projective Transformation

````python
# t3-projection.py - Projective transformation
# Projective matrix:
# [a  b  c]   [0.8    0.3    10  ]
# [d  e  f] = [-0.7   0.9    30  ]
# [g  h  1]   [0.0028 0.0036 1   ]
#
# KEY: Last row is NOT [0, 0, 1]!

sm = array([[.8, .3, 10],
            [-.7, .9, 30],
            [0.0028, 0.0036, 1]])

print("Projective Matrix:")
print(sm)

# Calculate output size (DIVIDE BY W!)
tp = sm @ array([[rows], [cols], [1]])
rowsd = int(tp[0, 0] / tp[2, 0]) + 150  # Divide by w!
colsd = int(tp[1, 0] / tp[2, 0]) + 150  # Divide by w!

# Create output image
im2_projective = zeros([rowsd, colsd])

# Forward mapping with perspective division
r = 0
while r < rows:
    c = 0
    while c < cols:
        tp = sm @ array([[r], [c], [1]])
        rd = tp[0, 0]
        cd = tp[1, 0]
        td = tp[2, 0]  # td is NOT always 1 for projective!
        
        # Convert homogeneous to Cartesian (CRITICAL!)
        rd = int(rd / td)
        cd = int(cd / td)
        td = int(td / td)
        
        if rd >= 0 and cd >= 0 and rd < rowsd and cd < colsd:
            im2_projective[rd, cd] = im1[r, c]
        c = c + 1
    r = r + 1

print("Projective transformation complete!")
````

````python
# Display projective result
figure(figsize=(12, 5))
subplot(1, 2, 1)
imshow(im1, cmap='gray')
title('Original')
subplot(1, 2, 2)
imshow(im2_projective, cmap='gray')
title('Projective')
show()
````

````python
# Save projective result
img2 = Image.fromarray(im2_projective.astype('uint8'), 'L')
img2.convert('RGB').save('projective_result.jpg', 'jpeg')
print("Saved as projective_result.jpg")
````

# SUMMARY: Key Differences

| Transformation | Matrix | Division by w? | Preserves |
|----------------|--------|----------------|------------|
| **Scale** | `[[sx,0],[0,sy]]` | No | Shape, parallel lines |
| **Rotation** | `[[cos,-sin,0],[sin,cos,0],[0,0,1]]` | No (w=1) | Shape, parallel lines |
| **Projective** | `[[a,b,c],[d,e,f],[g,h,1]]` | **YES!** | Only straight lines |
